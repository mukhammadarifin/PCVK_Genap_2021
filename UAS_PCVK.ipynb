```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(20, 10))
  
# setting values to rows and column variables
rows = 2
columns = 2
  
# reading images
img1 = cv2.imread('/content/drive/MyDrive/image/jokowi.jpg')
img2 = cv2.imread('/content/drive/MyDrive/image/peppers.jpg')
img3 = cv2.imread('/content/drive/MyDrive/image/parking-lot-cars.jpg')
img4 = cv2.imread('/content/drive/MyDrive/image/lily.jpg')

  
# Adds a subplot at the 1st position
fig.add_subplot(rows, columns, 1)
  
# showing image
plt.imshow(img1)
plt.axis('off')
plt.title("First")
  
# Adds a subplot at the 2nd position
fig.add_subplot(rows, columns, 2)
  
# showing image
plt.imshow(img2)
plt.axis('off')
plt.title("Second")

# Adds a subplot at the 3rd position
fig.add_subplot(rows, columns, 3)
  
# showing image
plt.imshow(img3)
plt.axis('off')
plt.title("Third")
  
# Adds a subplot at the 4th position
fig.add_subplot(rows, columns, 4)
  
# showing image
plt.imshow(img4)
plt.axis('off')
plt.title("Fourth")
```

```python
grid_RGB1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
grid_RGB2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
grid_RGB3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)
grid_RGB4 = cv2.cvtColor(img4, cv2.COLOR_BGR2RGB)
```

```python
fig = plt.figure(figsize=(20, 10))
  
# setting values to rows and column variables
rows = 2
columns = 2
  
# Adds a subplot at the 1st position
fig.add_subplot(rows, columns, 1)
  
# showing image
plt.imshow(grid_RGB1)
plt.axis('off')
plt.title("First")
  
# Adds a subplot at the 2nd position
fig.add_subplot(rows, columns, 2)
  
# showing image
plt.imshow(grid_RGB2)
plt.axis('off')
plt.title("Second")

# Adds a subplot at the 3rd position
fig.add_subplot(rows, columns, 3)
  
# showing image
plt.imshow(grid_RGB3)
plt.axis('off')
plt.title("Third")

# Adds a subplot at the 4th position
fig.add_subplot(rows, columns, 4)
  
# showing image
plt.imshow(grid_RGB4)
plt.axis('off')
plt.title("Fourth")
```

```python
grid_HSV1 = cv2.cvtColor(grid_RGB1, cv2.COLOR_RGB2HSV)
grid_HSV2 = cv2.cvtColor(grid_RGB2, cv2.COLOR_RGB2HSV)
grid_HSV3 = cv2.cvtColor(grid_RGB3, cv2.COLOR_RGB2HSV)
grid_HSV4 = cv2.cvtColor(grid_RGB4, cv2.COLOR_RGB2HSV)
```

```python
lower = np.array([0, 150, 50])
upper = np.array([10, 255, 255])
```

```python
mask1 = cv2.inRange(grid_HSV1, lower, upper)
mask2 = cv2.inRange(grid_HSV2, lower, upper)
mask3 = cv2.inRange(grid_HSV3, lower, upper)
mask4 = cv2.inRange(grid_HSV4, lower, upper)
```

```python
fig = plt.figure(figsize=(20, 10))
  
# setting values to rows and column variables
rows = 2
columns = 2
  
# Adds a subplot at the 1st position
fig.add_subplot(rows, columns, 1)
  
# showing image
plt.imshow(mask1)
plt.axis('off')
plt.title("First")
  
# Adds a subplot at the 2nd position
fig.add_subplot(rows, columns, 2)
  
# showing image
plt.imshow(mask2)
plt.axis('off')
plt.title("Second")

# Adds a subplot at the 3rd position
fig.add_subplot(rows, columns, 3)
  
# showing image
plt.imshow(mask3)
plt.axis('off')
plt.title("Third")

# Adds a subplot at the 4th position
fig.add_subplot(rows, columns, 4)
  
# showing image
plt.imshow(mask4)
plt.axis('off')
plt.title("Fourth")
```

```python
res1 = cv2.bitwise_and(grid_RGB1, grid_RGB1, mask=mask1)
res2 = cv2.bitwise_and(grid_RGB2, grid_RGB2, mask=mask2)
res3 = cv2.bitwise_and(grid_RGB3, grid_RGB3, mask=mask3)
res4 = cv2.bitwise_and(grid_RGB4, grid_RGB4, mask=mask4)
```

```python
fig = plt.figure(figsize=(30, 25))
  
# setting values to rows and column variables
rows = 4
columns = 2
  
# Adds a subplot at the 1st position
fig.add_subplot(rows, columns, 1)
  
# showing image
plt.imshow(grid_RGB1)
plt.axis('off')
plt.title("First")

# Adds a subplot at the 1st position
fig.add_subplot(rows, columns, 2)
  
# showing image
plt.imshow(res1)
plt.axis('off')
plt.title("First")
  
# Adds a subplot at the 2nd position
fig.add_subplot(rows, columns, 3)
  
# showing image
plt.imshow(grid_RGB2)
plt.axis('off')
plt.title("Second")

# Adds a subplot at the 2nd position
fig.add_subplot(rows, columns, 4)
  
# showing image
plt.imshow(res2)
plt.axis('off')
plt.title("Second")

# Adds a subplot at the 3rd position
fig.add_subplot(rows, columns, 5)
  
# showing image
plt.imshow(grid_RGB3)
plt.axis('off')
plt.title("Third")

# Adds a subplot at the 3rd position
fig.add_subplot(rows, columns, 6)
  
# showing image
plt.imshow(res3)
plt.axis('off')
plt.title("Third")

# Adds a subplot at the 4th position
fig.add_subplot(rows, columns, 7)
  
# showing image
plt.imshow(grid_RGB4)
plt.axis('off')
plt.title("Fourth")

# Adds a subplot at the 4th position
fig.add_subplot(rows, columns, 8)
  
# showing image
plt.imshow(res4)
plt.axis('off')
plt.title("Fourth")
```
